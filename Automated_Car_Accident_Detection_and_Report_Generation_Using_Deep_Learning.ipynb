```python
# Initialize a YOLOv11 model and train it on the 'accident' dataset.
# The training runs for 50 epochs with a batch size of 16 and image size of 640.
# Training results will be saved under 'accident_yolo11/exp1' project.
from ultralytics import YOLO

# Load YOLOv11 pretrained model
model = YOLO("yolo11n.pt")  # use small model for fast training

# Train
model.train(
    data="dataset/accident/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    device=0,
    patience=10,
    project="accident_yolo11",
    name="exp1"
)
```

```python
# Perform object detection using the trained YOLO model on a video file.
# Predictions are saved to 'runs/accident_output' with a confidence threshold of 0.4.
from ultralytics import YOLO

model = YOLO("/content/best (2).pt")

model.predict(
    source="/content/side_collision_123.mp4",
    conf=0.4,
    save=True,
    project="runs",
    name="accident_output",
    exist_ok=True
)
```

```python
# Perform object detection on the processed video and extract accident-related frames.
# For each frame with detections, record the frame index, number of detections, confidence scores, and bounding box coordinates.
from ultralytics import YOLO

model = YOLO("/content/best (2).pt")

results = model.predict(
    source="/content/runs/accident_output/side_collision_123.avi",
    conf=0.4,
    save=True,
    project="runs",
    name="accident_output",
    exist_ok=True
)

# logic to process frames
accident_frames = []
for i, r in enumerate(results):
    if len(r.boxes) > 0:  # If anything is detected
        # Extract metadata for reporting
        frame_data = {
            "frame_index": i,
            "detections": len(r.boxes),
            "confidence": [float(conf) for conf in r.boxes.conf],
            "boxes": r.boxes.xyxy.tolist()
        }
        accident_frames.append(frame_data)
```

```python
# Extract metrics from the accident detection results to create a data summary.
# Calculate max vehicles detected, average confidence, total frames with accidents, and assign a severity rating.
import json
from datetime import datetime

# Extract metrics for the NLP model
max_vehicles = max([len(r.boxes) for r in results]) if results else 0
avg_conf = sum([r.boxes.conf.mean().item() for r in results if len(r.boxes) > 0]) / len(results)
total_frames_with_accidents = sum(1 for r in results if len(r.boxes) > 0)

# Create a data summary for the NLP generator
data_summary = {
    "event": "Vehicle Collision",
    "peak_involvement": f"{max_vehicles} vehicles",
    "confidence_level": f"{avg_conf:.2%}",
    "duration_captured": f"{total_frames_with_accidents} frames",
    "severity": "High" if max_vehicles >= 3 else "Moderate"
}
```

```python
# Define a function to generate a narrative report from the data summary using natural language generation (NLG).
# Then, call the function to create the NLP narrative and print it.
def generate_nlp_report(summary):
    """
    Acts as a Natural Language Generation (NLG) engine to convert
    structured detection data into a narrative.
    """
    narrative = (
        f"Incident Analysis Report\n"
        f"-------------------------\n"
        f"On {datetime.now().strftime('%Y-%m-%d at %H:%M')}, the system identified a {summary['event']} "
        f"with a {summary['severity']} severity rating. \n\n"
        f"Technical Analysis:\n"
        f"Visual evidence indicates a peak involvement of {summary['peak_involvement']}. "
        f"The detection model maintained a confidence score of {summary['confidence_level']} throughout "
        f"the {summary['duration_captured']} of interest. \n\n"
        f"Conclusion:\n"
        f"The high overlap of bounding boxes and vehicle count suggests an immediate need for emergency response."
    )
    return narrative

# Generate the narrative
nlp_narrative = generate_nlp_report(data_summary)

# Display the NLP Report
print(nlp_narrative)
```

```python
# Combine the data summary and NLP narrative into a final report and save it as a JSON file.
import json
from datetime import datetime

final_report = {
    "metadata": data_summary,
    "narrative_report": nlp_narrative,
    "timestamp": datetime.now().isoformat()
}

with open('/content/accident_report.json', 'w') as f:
    json.dump(final_report, f, indent=4)

print("\n[SUCCESS] NLP Report generated and saved to /content/accident_report.json")
```
